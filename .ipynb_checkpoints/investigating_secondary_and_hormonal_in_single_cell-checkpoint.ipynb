```python
import pandas as pd
import numpy as np
import scanpy as sc
import h5py 
import anndata
import Go_annotations as go_an
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
single_cell_h5py = h5py.File('/data/passala/Data_from_CoCoCoNet/single_cell_data/Ara_data.hdf5','r')
list(single_cell_h5py.keys())
```

```python
col_data = single_cell_h5py['coldata']
row_data = single_cell_h5py['rowdata']
embedding_data = single_cell_h5py['embedding']
normalized_counts = single_cell_h5py['normalized_counts']
```

```python

row_data_decoded = []

for gene_name in row_data:
    row_data_decoded.append(gene_name[0].decode())

cell_type_number =[]
study_number = []
study_id = []
batch_cluster = []
meta_cluster = []

for cell_identity in col_data:
    cell_type_number.append(cell_identity[0])  
    study_number.append(cell_identity[1])
    study_id.append(cell_identity[2])
    batch_cluster.append(cell_identity[3])
    meta_cluster.append(cell_identity[4])

barcode_for_each_cell = [] 
for barcode in embedding_data:
    barcode_for_each_cell.append(barcode[2])
        
```

```python
len(barcode_for_each_cell)
```

```python
obs_arabidop = pd.DataFrame(index = barcode_for_each_cell, data = list(zip(cell_type_number,study_number,batch_cluster, meta_cluster)), columns = ['Cell Type','Study Number','Batch Cluster','Meta Cluster'] )

vars_arabidop = pd.DataFrame(index = row_data_decoded)
```

```python
single_cell_arabidopsis_root_4_datasets = anndata.AnnData(X = normalized_counts[:],obs = obs_arabidop, var = vars_arabidop)
# single_cell_arabidopsis_root_4_datasets.X.value = normalized_counts[:]
# single_cell_arabidopsis_root_4_datasets.X.shape()
```

```python
full_annotated_group_list = pd.read_csv('/data/passala/Generated_Tables/Reference_tables/all_annotated_go_groups.csv',encoding = 'cp1252',index_col= 0)
full_annotated_group_list.head(5)
```

```python
go_groups_of_interest = full_annotated_group_list.index.tolist()
```

```python

arabidopsis_annotations = go_an.get_go_annotations_as_panda('arabidopsis')
arabidopsis_annotations = arabidopsis_annotations[arabidopsis_annotations.columns.intersection(go_groups_of_interest)]
arabidopsis_annotations
```

```python
full_annotated_group_list.sort_values(by = 'Annotation')
```

```python
ribosomal_groups = full_annotated_group_list[full_annotated_group_list['Annotation'] == 'Ribosomal'].index.tolist()
hormonal_groups = full_annotated_group_list[full_annotated_group_list['Annotation'] == 'Hormonal'].index.tolist()
photosynthesis_groups = full_annotated_group_list[full_annotated_group_list['Annotation'] == 'Photosynthesis'].index.tolist()
secondary_groups = full_annotated_group_list[full_annotated_group_list['Annotation'] == 'Secondary Metabolism'].index.tolist()
```

```python
ribosomal_genes = []
for group in ribosomal_groups:
    genes_in_group = arabidopsis_annotations.loc[arabidopsis_annotations[group] == 1].index.tolist()
    ribosomal_genes.extend(genes_in_group)
ribosomal_genes = list(set(ribosomal_genes))
    
```

```python
hormonal_genes = []
for group in hormonal_groups:
    genes_in_group = arabidopsis_annotations.loc[arabidopsis_annotations[group] == 1].index.tolist()
    hormonal_genes.extend(genes_in_group)
hormonal_genes = list(set(hormonal_genes))
    
```

```python
photosynthesis_genes = []
for group in photosynthesis_groups:
    genes_in_group = arabidopsis_annotations.loc[arabidopsis_annotations[group] == 1].index.tolist()
    photosynthesis_genes.extend(genes_in_group)
photosynthesis_genes = list(set(photosynthesis_genes))
```

```python
secondary_genes = []
for group in secondary_groups:
    genes_in_group = arabidopsis_annotations.loc[arabidopsis_annotations[group] == 1].index.tolist()
    secondary_genes.extend(genes_in_group)
secondary_genes = list(set(secondary_genes))
```

```python
len(ribosomal_genes)
```

```python
list_of_tau_values = pd.read_csv('/home/passala/Gene_duplication_project/R_data_files/tau_values_arabidopsis_genes.csv',index_col=0)
tau_means = list_of_tau_values.mean(axis =1)
tau_means = pd.DataFrame(data = tau_means, columns = ['Tau Score'])
```

```python
tau_means  = tau_means.reset_index()
tau_means
```

```python
pd_ribosomal_genes = pd.DataFrame(data = ribosomal_genes, columns= ["Ribosome Genes"])
pd_ribosomal_genes['Ribosomal'] = 'Ribosomal' 
pd_ribosomal_genes
```

```python
pd_hormonal_genes = pd.DataFrame(data = hormonal_genes, columns= ["Hormonal Genes"])
pd_hormonal_genes['Hormonal'] = 'Hormonal'
pd_hormonal_genes
```

```python
pd_photosynthesis_genes = pd.DataFrame(data = photosynthesis_genes, columns= ["Photosynthesis Genes"])
pd_photosynthesis_genes['Photosynthesis'] = 'Photosynthesis'
pd_photosynthesis_genes
```

```python
pd_secondary_genes = pd.DataFrame(data = secondary_genes, columns= ["Secondary Metabolism Genes"])
pd_secondary_genes['Secondary Metabolism'] = 'Secondary Metabolism'
pd_secondary_genes
```

```python
tau_means = tau_means.merge(right = pd_ribosomal_genes, how ='left', left_on = 'Gene',right_on = 'Ribosome Genes')
tau_means = tau_means.merge(right = pd_hormonal_genes, how ='left', left_on = 'Gene',right_on = 'Hormonal Genes')
tau_means = tau_means.merge(right = pd_photosynthesis_genes, how ='left', left_on = 'Gene',right_on = 'Photosynthesis Genes')
tau_means = tau_means.merge(right = pd_secondary_genes, how ='left', left_on = 'Gene',right_on = 'Secondary Metabolism Genes')
tau_means
```

```python
tau_means = tau_means.drop(columns = ['Ribosome Genes','Hormonal Genes','Photosynthesis Genes','Secondary Metabolism Genes'])
tau_means['Unannotated'] = np.nan
tau_means
```

```python
tau_means
```

```python
tau_means['Unannotated'].loc[(tau_means['Ribosomal'] != 1) & (tau_means['Hormonal'] != 1) & (tau_means['Photosynthesis'] != 1) & (tau_means['Secondary Metabolism'] != 1)] = 'Unannotated'
```

```python
tau_means = tau_means.melt(id_vars= ['Gene','Tau Score'], var_name = "Gene Type", value_name = 'Present')
tau_means = tau_means.dropna()
tau_means
```

```python
tau_means_dropped_dupes = tau_means.drop_duplicates(subset = 'Gene')
tau_means_dropped_dupes
```

```python
my_pal = {'Ribosomal':'tab:purple','Hormonal':'tab:blue','Secondary Metabolism':'tab:red','Unannotated':'tab:gray'}
```

```python
fig, ax = plt.subplots(figsize = (20,15))
sns.boxplot(x ="Gene Type",y = 'Tau Score',data = tau_means.loc[tau_means['Gene Type']!= 'Photosynthesis'], palette= my_pal)
plt.ylabel("Tau Scores of Genes", fontsize = 25)
plt.xlabel("Gene Annotation", fontsize = 25)
#plt.title("As Intra-Set Co-expresssion Increases, the Range of Functional Conservation Values Increases", fontsize = 25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
```
